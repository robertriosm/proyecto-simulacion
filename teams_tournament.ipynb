## Combinations for the tournament's matches between teams

```python
import pandas as pd
```

#### cargar datos de la temporada 2023

```python
df = pd.read_csv('Team Comp vs Opp Comp_the big cheese_data.csv')
df = df[df["Season"] > 2022]
df
```

```python
team_names = set(df["team_name"])
team_names
```

```python
len(team_names)
```

```python
from itertools import combinations
teams_matches = list(combinations(team_names,r=2))
teams_matches
```
